```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Sine & Cosine
x = np.linspace(0, 4 * np.pi, 200)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, np.sin(x), label='sin(x)', color='#89b4fa')
axes[0].plot(x, np.cos(x), label='cos(x)', color='#f38ba8')
axes[0].set_title('Trigonometric Functions')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Random scatter
np.random.seed(42)
n = 100
axes[1].scatter(
    np.random.randn(n),
    np.random.randn(n),
    c=np.random.rand(n),
    s=np.random.rand(n) * 200,
    alpha=0.6,
    cmap='coolwarm',
)
axes[1].set_title('Random Scatter')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
