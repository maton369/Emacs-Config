```
import numpy as np
import matplotlib.pyplot as plt
```

```
# 1. Trigonometric Functions & Scatter
x = np.linspace(0, 4 * np.pi, 200)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, np.sin(x), label='sin(x)', color='#89b4fa')
axes[0].plot(x, np.cos(x), label='cos(x)', color='#f38ba8')
axes[0].set_title('Trigonometric Functions')
axes[0].legend()
axes[0].grid(alpha=0.3)

np.random.seed(42)
n = 100
axes[1].scatter(
    np.random.randn(n), np.random.randn(n),
    c=np.random.rand(n), s=np.random.rand(n) * 200,
    alpha=0.6, cmap='coolwarm',
)
axes[1].set_title('Random Scatter')
axes[1].grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```
# 2. Bar Chart & Histogram
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

categories = ['Python', 'JS', 'Rust', 'Go', 'Java', 'C++']
values = [92, 78, 65, 58, 72, 55]
colors = ['#89b4fa', '#f9e2af', '#fab387', '#a6e3a1', '#f38ba8', '#cba6f7']
axes[0].bar(categories, values, color=colors, edgecolor='white', linewidth=0.5)
axes[0].set_title('Language Popularity')
axes[0].set_ylabel('Score')

data = np.random.exponential(scale=2, size=1000)
axes[1].hist(data, bins=40, color='#89b4fa', alpha=0.7, edgecolor='white')
axes[1].axvline(data.mean(), color='#f38ba8', linestyle='--', label=f'Mean={data.mean():.1f}')
axes[1].set_title('Exponential Distribution')
axes[1].legend()
plt.tight_layout()
plt.show()
```

```
# 3. Heatmap & Pie Chart
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

np.random.seed(0)
corr = np.corrcoef(np.random.randn(6, 50))
im = axes[0].imshow(corr, cmap='RdBu_r', vmin=-1, vmax=1)
axes[0].set_title('Correlation Heatmap')
for i in range(6):
    for j in range(6):
        axes[0].text(j, i, f'{corr[i,j]:.2f}', ha='center', va='center', fontsize=8)
fig.colorbar(im, ax=axes[0], shrink=0.8)

sizes = [35, 25, 20, 12, 8]
labels = ['Web', 'Data', 'Mobile', 'DevOps', 'Other']
explode = (0.05, 0, 0, 0, 0)
axes[1].pie(sizes, explode=explode, labels=labels, autopct='%1.0f%%',
            colors=['#89b4fa', '#a6e3a1', '#f9e2af', '#f38ba8', '#cba6f7'],
            startangle=90, wedgeprops={'edgecolor': 'white'})
axes[1].set_title('Dev Focus Areas')
plt.tight_layout()
plt.show()
```

```
# 4. Time Series & Area Chart
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

np.random.seed(7)
days = np.arange(365)
trend = 50 + 0.05 * days + 10 * np.sin(2 * np.pi * days / 365)
noise = np.cumsum(np.random.randn(365) * 0.5)
price = trend + noise
axes[0].plot(days, price, color='#89b4fa', linewidth=0.8)
axes[0].fill_between(days, price, alpha=0.15, color='#89b4fa')
ma30 = np.convolve(price, np.ones(30)/30, mode='valid')
axes[0].plot(days[29:], ma30, color='#f38ba8', linewidth=1.5, label='30-day MA')
axes[0].set_title('Simulated Stock Price')
axes[0].legend()
axes[0].set_xlabel('Day')

x = np.linspace(0, 10, 200)
y1 = np.exp(-0.1 * x) * np.sin(x) ** 2
y2 = np.exp(-0.1 * x) * np.cos(x) ** 2
axes[1].fill_between(x, 0, y1, alpha=0.6, color='#89b4fa', label='sin²')
axes[1].fill_between(x, y1, y1 + y2, alpha=0.6, color='#f38ba8', label='cos²')
axes[1].set_title('Stacked Area (Damped Oscillation)')
axes[1].legend()
plt.tight_layout()
plt.show()
```

```
# 5. Box Plot & Violin Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

np.random.seed(3)
data_box = [np.random.normal(loc, 1.5, 200) for loc in [2, 3, 5, 4, 6]]
bp = axes[0].boxplot(data_box, patch_artist=True, labels=['A', 'B', 'C', 'D', 'E'])
box_colors = ['#89b4fa', '#a6e3a1', '#f9e2af', '#f38ba8', '#cba6f7']
for patch, color in zip(bp['boxes'], box_colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[0].set_title('Box Plot')
axes[0].grid(axis='y', alpha=0.3)

vp = axes[1].violinplot(data_box, showmeans=True, showmedians=True)
for i, body in enumerate(vp['bodies']):
    body.set_facecolor(box_colors[i])
    body.set_alpha(0.7)
axes[1].set_title('Violin Plot')
axes[1].set_xticks([1, 2, 3, 4, 5])
axes[1].set_xticklabels(['A', 'B', 'C', 'D', 'E'])
axes[1].grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```
# 6. Polar Plot & Contour
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(121, projection='polar')
theta = np.linspace(0, 2 * np.pi, 100)
r1 = 1 + np.cos(3 * theta)
r2 = 1 + np.sin(5 * theta)
ax1.plot(theta, r1, color='#89b4fa', linewidth=2, label='r=1+cos(3θ)')
ax1.plot(theta, r2, color='#f38ba8', linewidth=2, label='r=1+sin(5θ)')
ax1.set_title('Polar Roses', pad=15)
ax1.legend(loc='lower right', fontsize=8)

ax2 = fig.add_subplot(122)
x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2) * np.exp(-(X**2 + Y**2) / 6)
cs = ax2.contourf(X, Y, Z, levels=20, cmap='viridis')
ax2.contour(X, Y, Z, levels=20, colors='white', linewidths=0.3, alpha=0.5)
fig.colorbar(cs, ax=ax2, shrink=0.8)
ax2.set_title('Contour: sin(x²+y²)·exp(-(x²+y²)/6)')
ax2.set_aspect('equal')
plt.tight_layout()
plt.show()
```
